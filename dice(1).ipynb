```python
n = int(input("How many times do you want to roll the dice?"))
```

```python
import random

dice = []
for i in range(0,6):
    dice.append(0)

for i in range(n):
    andis = random.randrange(0,6)
    dice[andis] =  dice[andis] + 1
dice
```

```python
import matplotlib.pyplot as plt

plt.bar(range(1,7), dice)
```
